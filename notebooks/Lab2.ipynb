```python
# Импортируем необходимые библиотеки
import pickle
import numpy as np
import matplotlib.pyplot as plt

# 1. Функция для открытия файлов CIFAR-100
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='latin1')
    return dict

# 2. Указываем путь к вашим файлам CIFAR-100
train_file = 'cifar-100-python/train'
meta_file = 'cifar-100-python/meta'

# 3. Загружаем данные
train_data = unpickle(train_file)
meta_data = unpickle(meta_file)

# 4. Задаем ваши классы по варианту
MY_CLASSES = [10, 39, 59]  # bowl, keyboard, pine_tree
CLASS_NAMES = {
    10: 'bowl (миска)',
    39: 'keyboard (клавиатура)',
    59: 'pine_tree (сосна)'
}
# Получаем список всех названий классов из мета-файла (на всякий случай)
all_fine_names = meta_data['fine_label_names']

print("="*50)
print("Ваши классы:")
for class_id in MY_CLASSES:
    print(f"  ID: {class_id} -> {CLASS_NAMES[class_id]}")
print("="*50)

# 5. Извлекаем данные и метки
X_train_raw = train_data['data']  # Все изображения (50000, 3072)
y_train_raw = np.array(train_data['fine_labels'])  # Все метки (50000)

# 6. Создаем маску для отбора ТОЛЬКО ваших классов
mask = np.isin(y_train_raw, MY_CLASSES)

# 7. Применяем маску к данным и меткам
X_train_filtered = X_train_raw[mask]
y_train_filtered = y_train_raw[mask]

print(f"Всего изображений выбранных классов: {len(X_train_filtered)}")
for class_id in MY_CLASSES:
    count = np.sum(y_train_filtered == class_id)
    print(f"  Класс {CLASS_NAMES[class_id]}: {count} изображений")

# 8. Преобразуем отфильтрованные изображения в правильный формат (N, 32, 32, 3)
X_train_reshaped = X_train_filtered.reshape(-1, 3, 32, 32)
X_train_display = X_train_reshaped.transpose(0, 2, 3, 1)

print("Форма данных для отображения:", X_train_display.shape)

# 9. Функция для отображения по 5 изображений каждого класса
def show_5_per_class(images, labels, class_ids, class_names, images_per_class=5):
    """
    Отображает по `images_per_class` изображений для каждого класса.
    Всего получается len(class_ids) * images_per_class картинок,
    расположенных в rows = len(class_ids) рядов по cols = images_per_class столбцов.
    """
    rows = len(class_ids)
    cols = images_per_class

    fig, axes = plt.subplots(rows, cols, figsize=(cols*2.5, rows*2.5))
    fig.suptitle(f"По {images_per_class} изображений каждого класса", fontsize=16, y=1.02)

    for row, class_id in enumerate(class_ids):
        # Находим все индексы изображений текущего класса
        class_indices = np.where(labels == class_id)[0]

        # Выбираем случайные images_per_class индексов
        selected_indices = np.random.choice(class_indices, images_per_class, replace=False)

        for col, idx in enumerate(selected_indices):
            ax = axes[row, col]
            ax.imshow(images[idx])
            ax.set_title(f"{class_names[class_id]}", fontsize=9)
            ax.axis('off')

    plt.tight_layout()
    plt.show()

# 10. Вызываем функцию
show_5_per_class(X_train_display,
                 y_train_filtered,
                 MY_CLASSES,
                 CLASS_NAMES,
                 images_per_class=5)
```

## Создание Pytorch модели сверточной нейронной сети

```python
class Normalize(nn.Module):
    def __init__(self, mean, std):
        super(Normalize, self).__init__()
        self.mean = torch.tensor(mean).to(device)
        self.std = torch.tensor(std).to(device)

    def forward(self, input):
        x = input / 255.0
        x = x - self.mean
        x = x / self.std
        return x.permute(0, 3, 1, 2) # nhwc -> nm

class GlobalMaxPool2d(nn.Module):
    def __init__(self):
        super(GlobalMaxPool2d, self).__init__()

    def forward(self, input):
        out = F.adaptive_max_pool2d(input, output_size=1)
        return out.flatten(start_dim=1)
```

```python
class Cifar100_CNN(nn.Module):
    def __init__(self, hidden_size=32, classes=100):
        super(Cifar100_CNN, self).__init__()
        self.seq = nn.Sequential(
            Normalize([0.5074, 0.4867, 0.4411], [0.2011, 0.1987, 0.2025]),

            # Первый сверточный слой: уменьшаем размерность через stride=2
            nn.Conv2d(3, hidden_size, kernel_size=5, stride=2, padding=2),
            nn.ReLU(),

            # Второй сверточный слой: еще уменьшаем через stride=2
            nn.Conv2d(hidden_size, hidden_size*2, kernel_size=3, stride=2, padding=1),
            nn.ReLU(),

            # Вместо пуллинга используем адаптивный пуллинг для приведения к размеру 1x1
            nn.AdaptiveAvgPool2d((1, 1)),

            nn.Flatten(),
            nn.Linear(hidden_size*2, classes),
        )

    def forward(self, input):
        return self.seq(input)

# Создаем модель
HIDDEN_SIZE = 32
model = Cifar100_CNN(hidden_size=HIDDEN_SIZE, classes=3)
model.to(device)

# Выводим summary
print(model(torch.rand(1, 32, 32, 3).to(device)))
summary(model, input_size=(32, 32, 3))
```

#### 2. Изменение размера ядра свертки (5→3)


```python
class Cifar100_CNN(nn.Module):
    def __init__(self, hidden_size=32, classes=100):
        super(Cifar100_CNN, self).__init__()
        self.seq = nn.Sequential(
            Normalize([0.5074, 0.4867, 0.4411], [0.2011, 0.1987, 0.2025]),

            # Ядро 3x3 (было 5x5)
            nn.Conv2d(3, hidden_size, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(4),

            nn.Conv2d(hidden_size, hidden_size*2, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(4),

            nn.Flatten(),
            nn.Linear(hidden_size*2 * 2 * 2, classes),
        )

    def forward(self, input):
        return self.seq(input)

# Создание модели
HIDDEN_SIZE = 32
model = Cifar100_CNN(hidden_size=HIDDEN_SIZE, classes=3)
model.to(device)
summary(model, input_size=(32, 32, 3))
```

#### 3. Изменение шага свёртки (stride=1→2)


```python
class Cifar100_CNN(nn.Module):
    def __init__(self, hidden_size=32, classes=100):
        super(Cifar100_CNN, self).__init__()
        self.seq = nn.Sequential(
            Normalize([0.5074, 0.4867, 0.4411], [0.2011, 0.1987, 0.2025]),

            # Шаг 2 вместо 1
            nn.Conv2d(3, hidden_size, kernel_size=5, stride=2, padding=2),  # 32×32 → 16×16
            nn.ReLU(),
            nn.MaxPool2d(4),  # 16×16 → 4×4

            nn.Conv2d(hidden_size, hidden_size*2, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(4),  # 4×4 → 1×1

            nn.Flatten(),
            # Размер для Linear изменился!
            nn.Linear(hidden_size*2 * 1 * 1, classes),  # 64 * 1 * 1 = 64
        )

    def forward(self, input):
        return self.seq(input)

# Создание модели
HIDDEN_SIZE = 32
model = Cifar100_CNN(hidden_size=HIDDEN_SIZE, classes=3)
model.to(device)
summary(model, input_size=(32, 32, 3))
```

#### 4. Добавление третьего сверточного слоя


```python
class Cifar100_CNN(nn.Module):
    def __init__(self, hidden_size=32, classes=100):
        super(Cifar100_CNN, self).__init__()
        self.seq = nn.Sequential(
            Normalize([0.5074, 0.4867, 0.4411], [0.2011, 0.1987, 0.2025]),

            # Первый блок (с меньшим пуллингом)
            nn.Conv2d(3, hidden_size, kernel_size=5, stride=1, padding=2),
            nn.ReLU(),
            nn.MaxPool2d(2),  # 32×32 → 16×16

            # Второй блок
            nn.Conv2d(hidden_size, hidden_size*2, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),  # 16×16 → 8×8

            # Третий блок (НОВЫЙ)
            nn.Conv2d(hidden_size*2, hidden_size*4, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),  # 8×8 → 4×4

            nn.Flatten(),
            # Размер для Linear: 128 * 4 * 4 = 2048
            nn.Linear(hidden_size*4 * 4 * 4, classes),
        )

    def forward(self, input):
        return self.seq(input)

# Создание модели
HIDDEN_SIZE = 32
model = Cifar100_CNN(hidden_size=HIDDEN_SIZE, classes=3)
model.to(device)
summary(model, input_size=(32, 32, 3))
```

#### 5. Добавление Dropout


```python
class Cifar100_CNN(nn.Module):
    def __init__(self, hidden_size=32, classes=100, dropout_prob=0.5):
        super(Cifar100_CNN, self).__init__()
        self.seq = nn.Sequential(
            Normalize([0.5074, 0.4867, 0.4411], [0.2011, 0.1987, 0.2025]),

            nn.Conv2d(3, hidden_size, kernel_size=5, stride=1, padding=2),
            nn.ReLU(),
            nn.MaxPool2d(4),

            nn.Conv2d(hidden_size, hidden_size*2, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(4),

            nn.Flatten(),
            nn.Dropout(dropout_prob),  # Dropout перед Linear
            nn.Linear(hidden_size*2 * 2 * 2, classes),
        )

    def forward(self, input):
        return self.seq(input)

# Создание модели
HIDDEN_SIZE = 32
model = Cifar100_CNN(hidden_size=HIDDEN_SIZE, classes=3, dropout_prob=0.5)
model.to(device)
summary(model, input_size=(32, 32, 3))
```

```python
EPOCHS = 80
REDRAW_EVERY = 20
steps_per_epoch = len(dataloader['train'])
steps_per_epoch_val = len(dataloader['test'])
# NEW
pbar = tqdm(total=EPOCHS*steps_per_epoch)
losses = []
losses_val = []
passed = 0
for epoch in range(EPOCHS):  # проход по набору данных несколько раз
    #running_loss = 0.0
    tmp = []
    model.train()
    for i, batch in enumerate(dataloader['train'], 0):
        # получение одного минибатча; batch это двуэлементный список из [inputs, labels]
        inputs, labels = batch
        # на GPU
        inputs, labels = inputs.to(device), labels.to(device)

        # очищение прошлых градиентов с прошлой итерации
        optimizer.zero_grad()

        # прямой + обратный проходы + оптимизация
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        #loss = F.cross_entropy(outputs, labels)
        loss.backward()
        optimizer.step()

        # для подсчёта статистик
        #running_loss += loss.item()
        accuracy = (labels.detach().argmax(dim=-1)==outputs.detach().argmax(dim=-1)).\
                    to(torch.float32).mean().cpu()*100
        tmp.append((loss.item(), accuracy.item()))
        pbar.update(1)
    #print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / steps_per_epoch:.3f}')
    losses.append((np.mean(tmp, axis=0),
                   np.percentile(tmp, 25, axis=0),
                   np.percentile(tmp, 75, axis=0)))
    #running_loss = 0.0
    tmp = []
    model.eval()
    with torch.no_grad(): # отключение автоматического дифференцирования
        for i, data in enumerate(dataloader['test'], 0):
            inputs, labels = data
            # на GPU
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            #running_loss += loss.item()
            accuracy = (labels.argmax(dim=-1)==outputs.argmax(dim=-1)).\
                        to(torch.float32).mean().cpu()*100
            tmp.append((loss.item(), accuracy.item()))
    #print(f'[{epoch + 1}, {i + 1:5d}] val loss: {running_loss / steps_per_epoch_val:.3f}')
    losses_val.append((np.mean(tmp, axis=0),
                       np.percentile(tmp, 25, axis=0),
                       np.percentile(tmp, 75, axis=0)))
    if (epoch+1) % REDRAW_EVERY != 0:
        continue
    clear_output(wait=False)
    passed += pbar.format_dict['elapsed']
    pbar = tqdm(total=EPOCHS*steps_per_epoch, miniters=5)
    pbar.update((epoch+1)*steps_per_epoch)
    x_vals = np.arange(epoch+1)
    _, ax = plt.subplots(1, 2, figsize=(15, 5))
    stats = np.array(losses)
    stats_val = np.array(losses_val)
    ax[1].set_ylim(stats_val[:, 0, 1].min()-5, 100)
    ax[1].grid(axis='y')
    for i, title in enumerate(['CCE', 'Accuracy']):
        ax[i].plot(x_vals, stats[:, 0, i], label='train')
        ax[i].fill_between(x_vals, stats[:, 1, i],
                           stats[:, 2, i], alpha=0.4)
        ax[i].plot(x_vals, stats_val[:, 0, i], label='val')
        ax[i].fill_between(x_vals,
                           stats_val[:, 1, i],
                           stats_val[:, 2, i], alpha=0.4)
        ax[i].legend()
        ax[i].set_title(title)
    plt.show()
print('Обучение закончено за %s секунд' % passed)
```

# Loss landscape

## идея https://arxiv.org/abs/1712.09913

```python
from copy import deepcopy
state_dict_backup = deepcopy(model.state_dict())
```

```python
def generate_theta(seed=None):
    model.eval()
    model.load_state_dict(state_dict_backup)
    if seed is not None:
        np.random.seed(seed)
    params = []
    with torch.no_grad():
        for w in model.parameters():
            params.append(w.detach().cpu().numpy())
    params_n = np.concatenate([p.flatten() for p in params]).size
    random_theta_flat = np.random.normal(size=params_n)
    random_theta = []
    offset = 0
    for p in params:
        p_weights = p.flatten().size
        random_theta.append(random_theta_flat[offset:offset+p_weights].\
                            reshape(*p.shape))
        # normalization
        rank = random_theta[-1].shape.__len__()
        if rank == 4: # convolution
            #print('Conv')
            # Frobenius norm
            norm_r = np.sqrt((random_theta[-1]**2).sum(axis=-1).sum(axis=-1))
            norm_p = np.sqrt((p**2).sum(axis=-1).sum(axis=-1))
            norm = (norm_p / norm_r).reshape(*norm_p.shape, 1, 1)
        elif rank == 2: # fully connected
            #print('FC')
            norm_r = np.sqrt((random_theta[-1]**2).sum(axis=-1))
            norm_p = np.sqrt((p**2).sum(axis=-1))
            norm = (norm_p / norm_r).reshape(-1, 1)
        elif rank == 1: # bias
            #print('bias')
            norm_r = np.sqrt((random_theta[-1]**2).sum())
            norm_p = np.sqrt((p**2).sum())
            norm = norm_p / norm_r
        random_theta[-1] = random_theta[-1]*norm
        offset += p_weights
    assert offset==params_n, \
            "Not all params are utilized. Expected %d, found %d"%(params_n, offset)
    return random_theta

theta1 = generate_theta(seed=0)
theta2 = generate_theta(seed=1011)
```

```python
criterion2 = nn.CrossEntropyLoss(reduction='none')
loss_curve = []
alphas = np.arange(-1500, 1500, step=5)/1000
for alpha in tqdm(alphas):
    state_dict = model.state_dict()
    for (k, v), v_new in zip(state_dict_backup.items(), theta1):
        tensor = v.clone().detach() + \
              alpha*(torch.tensor(v_new, device=device) - v.clone().detach())
        state_dict[k] = tensor
    model.load_state_dict(state_dict)
    loss = []

    with torch.no_grad(): # отключение автоматического дифференцирования
        for i, data in enumerate(dataloader['test'], 0):
            inputs, labels = data
            # на GPU
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs).detach()
            loss.append(criterion2(outputs, labels).detach().cpu().numpy())
    loss_curve.append(np.concatenate(loss).mean())
```

```python
Z = []
for a1, a2 in tqdm(alphas):
    state_dict = model.state_dict()
    for (k, v), v_new, v_new2 in zip(state_dict_backup.items(), theta1, theta2):
        # линейная интерполяция
        tensor = v.clone().detach() + \
                 a1*(torch.tensor(v_new, device=device) - \
                     v.clone().detach())
        # ещё раз со вторым вектором
        tensor = tensor + a2*(torch.tensor(v_new2, device=device) - tensor)
        state_dict[k] = tensor
        #print(k, tensor, v_new)
    model.load_state_dict(state_dict)
    loss = []
    with torch.no_grad(): # отключение автоматического дифференцирования
        for i, data in enumerate(dataloader['test'], 0):
            inputs, labels = data
            # на GPU
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs).detach()#.cpu().numpy()
            loss.append(criterion2(outputs, labels).detach().cpu().numpy())
    Z.append(np.concatenate(loss).mean())
ZZ = np.array(Z)
```
